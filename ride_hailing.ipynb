```python
# created a sorted df by city and week
df.sort_values(['City Name', 'Week Start'], inplace = True)
```

# Demand metrics calculation

```python
# peak session vs none-peak session ratio:
df['requesting_sessions_peak_ratio'] = df.requesting_sessions_peak / df.requesting_sessions
```

```python
# requesting session growth by cities
a = df.sort_values(['City Name', 'Week Start']).groupby('City Name')['requesting_sessions', 'requesting_sessions_peak'].pct_change().rename(
    columns={'requesting_sessions':'requesting_sessions_growth', 'requesting_sessions_peak':'requesting_sessions_peak_growth'})
```

```python
# add growth back to the df
df['requesting_sessions_growth'], df['peak_requesting_sessions_growth'] = a['requesting_sessions_growth'], a['requesting_sessions_peak_growth']
```

```python
# riders not getting serviced due to high demand or no supply

df['rider_stressed_zero'] = (df['non_requesting_stressed_sessions'] + df['zero_sessions']) / df['non_requesting_sessions']
df['peak_rider_stressed_zero'] = (df['peak_stressed_sessions'] + df['peak_zero_sessions']) / df['peak_non_requesting_sessions']
```

# Supply metrics calculation

```python
# en_route hours out of supply hours pct

df['en_route_trip_ratio'] = df.en_route_hours / df.on_trip_hours
df['peak_en_trip_hours_ratio'] = df.peak_en_route_hours / df.peak_on_trip_hours
```

```python
# supply hours growth by cities
b = df.sort_values(['City Name', 'Week Start']).groupby('City Name')['supply_hours', 'peak_supply_hours'].pct_change().rename(
    columns={'supply_hours':'supply_hours_growth', 'peak_supply_hours':'peak_supply_hours_growth'})
```

```python
# add growth back to the df
df['supply_hours_growth'], df['peak_supply_hours_growth'] = b['supply_hours_growth'], b['peak_supply_hours_growth']
```

# Matching efficiency

```python
# completed trips out of requesting sessions

df['trip_completed_ratio'] = df.completed_trips / df.requesting_sessions
df['peak_trip_complete_ratio'] = df.completed_peak / df.requesting_sessions_peak
df['cancel_rs'] = df.rider_cancel_rs + df.driver_cancel_rs
```

```python
key_metrics = ['City Name', 'requesting_sessions_peak_ratio', 'requesting_sessions', 'requesting_sessions_growth', 
               'supply_hours', 'supply_hours_growth', 'rider_stressed_zero', 'rider_cancel_rs', 'driver_cancel_rs',
               'cancel_rs', 'en_route_trip_ratio', 'tph', 'trip_completed_ratio']
```

```python
# create metrics table
df_metrics = df[key_metrics]
```

```python
df_metrics.head()
```

```python
# calculate mean for each metric by city
metrics_table = df_metrics.groupby('City Name').mean().reset_index()
metrics_table
```

```python
# rank city by city metric. Note for some metrics are higher the better but some others the lower the better
metrics_table['requesting_sessions_ranked'] = metrics_table['requesting_sessions'].rank(ascending=False)
metrics_table['requesting_sessions_growth_ranked'] = metrics_table['requesting_sessions_growth'].rank(ascending=False)
metrics_table['supply_hours_ranked'] = metrics_table['supply_hours'].rank(ascending=False)
metrics_table['supply_hours_growth_ranked'] = metrics_table['supply_hours_growth'].rank(ascending=False)
metrics_table['rider_stressed_zero_ranked'] = metrics_table['rider_stressed_zero'].rank(ascending=True)
metrics_table['rider_cancel_rs_ranked'] = metrics_table['rider_cancel_rs'].rank(ascending=True)
metrics_table['driver_cancel_rs_ranked'] = metrics_table['driver_cancel_rs'].rank(ascending=True)
metrics_table['cancel_rs_ranked'] = metrics_table['cancel_rs'].rank(ascending=True)
metrics_table['en_route_trip_ratio_ranked'] = metrics_table['en_route_trip_ratio'].rank(ascending=True)
metrics_table['tph_ranked'] = metrics_table['tph'].rank(ascending=False)
metrics_table['trip_completed_ratio_ranked'] = metrics_table['trip_completed_ratio'].rank(ascending=False)
metrics_table['requesting_sessions_peak_ratio_ranked'] = metrics_table['requesting_sessions_peak_ratio'].rank(ascending=True)
metrics_table_ranked = pd.concat([metrics_table.loc[:, 'City Name'], metrics_table.loc[:, 'requesting_sessions_ranked': 'requesting_sessions_peak_ratio_ranked']], axis=1)
```

```python
# preview ranking table
metrics_table_ranked
```

```python
# weight metrics as there are only 4 metrics size metrics vs 8 efficiency metrics
metrics_table_ranked['Weighted Total Rank Score'] = metrics_table_ranked.sum(axis=1) + metrics_table_ranked.loc[:, 'requesting_sessions_ranked':
                                                                                    'supply_hours_growth_ranked'].sum(axis=1)
metrics_table_ranked['Total Rank'] = metrics_table_ranked['Weighted Total Rank Score'].rank(ascending=True)

# preview the ranking table
metrics_table_ranked
```
